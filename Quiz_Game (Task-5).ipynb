# QUIZ GAME

```python
import random

# Create a list of quiz questions in the format (question, correct answer, options)
quiz_questions = [
    ("What is the capital of India?", "New Delhi", ["New Delhi", "Mumbai", "Kolkata", "Chennai"]),
    ("What is the square root of 144?", "12", ["21" , "14" , "12" , "22"]),
    ("Which gas do plants absorb from the atmosphere?", "Carbon dioxide", [ "Carbohydrate" ,  "Carbon dioxide" ,  "oxygen" ,  "nitrogen"]),
    ("What is the chemical symbol for gold?", "Au", ["su" , "gl" , "g" ,"Au"]),
    ("How many continents are there on Earth?", "7", ["7" , "5" , "8" ,"6"]),
]

# to show a question and its options 
def show_question(question, choices):
    print(question)
    if choices:
        for i, choice in enumerate(choices, start=1):
            print(f"{i}. {choice}")

# to get answer
def answer(choices):
    while True:
        try:
            user_answer = int(input("Enter your answer: "))
            if 1 <= user_answer <= len(choices):
                return choices[user_answer - 1]
            else:
                print("Invalid choice. Please enter a valid number.")
        except ValueError:
            print("Invalid input. Please enter a number.")

# to play the quiz
def play_quiz(questions):
    score = 0

    print("WELCOME TO THE QUIZ GAME!")
    print("You will be presented with multiple-choice questions.")
    print("Type the number corresponding to your answer.")

    random.shuffle(questions)

    for question, correct_answer, choices in questions:
        show_question(question, choices)
        user_answer = answer(choices)

        if user_answer == correct_answer:
            print("Correct!\n")
            score += 1
        else:
            print(f"Sorry, the correct answer is '{correct_answer}'.\n")

    print(f"Quiz completed! Your score: {score}/{len(questions)}")
    if score == len(questions):
        print("Congratulations! You got all questions correct.")
    elif score >= len(questions) / 2:
        print("Good! You did well.")
    else:
        print("You can do better. Keep practicing!")

if __name__ == "__main__":
    play_quiz(quiz_questions)
```
