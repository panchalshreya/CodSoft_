# TO DO LIST

```python
# Empty dictionary to store Values
to_do_list = {}

# Using def function and adding input function 
def add_task():
    list_name = input("Enter task name: ")
    to_do_list[list_name] = False
    print(f"Task '{list_name}' added!")

# Using def function to list tasks
def list_tasks():
    if to_do_list:
        print("Tasks:")
        for task, completed in to_do_list.items():
            show = "Done" if completed else "Not Done"
            print(f"- {task} ({show})")
    else:
        print("No tasks found!")

# For mark a task as done
def mark_task_done():
    list_name = input("Enter task name to mark as done: ")
    if list_name in to_do_list:
        to_do_list[list_name] = True
        print(f"Task '{list_name}' marked as done!")
    else:
        print(f"Task '{list_name}' not found!")

# For delete a task
def delete_task():
    list_name = input("Enter task name to delete: ")
    if list_name in to_do_list:
        del to_do_list[list_name]
        print(f"Task '{list_name}' deleted!")
    else:
        print(f"Task '{list_name}' not found!")

# Using loop funtion 
while True:
    print("\nTo Do List:")
    print("1. Add Task")
    print("2. List of Tasks")
    print("3. Mark Task as Done")
    print("4. Delete Task")
    print("5. Quit")
    
    select = input("Enter your choice: ")
    
    if select == "1":
        add_task()
    elif select == "2":
        list_tasks()
    elif select == "3":
        mark_task_done()
    elif select == "4":
        delete_task()
    elif select == "5":
        break
    else:
        print("Invalid choice. Please try again.")
```
